```
import numpy as np
import tensorflow as tf
import tensorflow_addons as tfa
import tensorflow_datasets as tfds
from matplotlib import pyplot as plt
```

```
image = tf.image.decode_jpeg(
    tf.io.read_file("./26172649646_931ea3cffc_b.jpg")
)
image.shape
```

```
plt.imshow(image)
plt.title("Original Image")
plt.show()
```

```
contrast_image = tf.image.adjust_contrast(image, 2.5)
plt.imshow(contrast_image)
plt.title("High Contrast")
plt.show()
```

```
bright_image = tf.image.adjust_brightness(image, 0.8)
plt.imshow(bright_image)
plt.title("Bright Image")
plt.show()
```

```
sharp_image = tfa.image.sharpness(image, 20)
plt.imshow(sharp_image)
plt.title("Sharp Image")
plt.show()
```

```
hue_image = tf.image.adjust_hue(image, 1.5)
plt.imshow(hue_image)
plt.title("Increased Hue")
plt.show()
```

```
sat_image = tf.image.adjust_saturation(image, 5.5)
plt.imshow(sat_image)
plt.title("High Saturated")
plt.show()
```

```
hist = np.bincount(image.numpy().flatten(), minlength=256)
plt.bar(np.arange(256), hist)
plt.title("Histogram of Original Image")
plt.show()
```

```
eq_image = tfa.image.equalize(image)
plt.imshow(eq_image)
plt.title("Histogram Equalized Image")
plt.show()
```

```
hist = np.bincount(eq_image.numpy().flatten(), minlength=256)
plt.bar(np.arange(256), hist)
plt.title("Histogram of Equalized Image")
plt.show()
```

```
rot_image = tfa.image.rotate(image, 45)
plt.imshow(rot_image)
plt.title("Image Rotated by 45°")
plt.show()
```

```
trans_image = tfa.image.translate_xy(image, [300, 200], (0, 0, 0))
plt.imshow(trans_image)
plt.title("Image Translated X by 300 and Y by 200")
plt.show()
```
