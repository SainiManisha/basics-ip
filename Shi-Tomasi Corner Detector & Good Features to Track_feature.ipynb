```python
import numpy as np
import cv2 #OpenCV-Python
%matplotlib inline
import matplotlib.pyplot as plt

print("OpenCV-Python Version %s" % cv2.__version__)
```

```python
#Import an image
image = cv2.imread('C:/Users/Manisha/Desktop/OpenCV/images/wheel.png')
```

```python
plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
```

```python
# Convert image color(BGR->Grayscale)
gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
# You can convert the image when calling cv2.imread()
# gray = cv2.imread('images/beaver.png', cv2.IMREAD_GRAYSCALE)

print(str(image.shape) + " => " + str(gray.shape))
plt.imshow(gray, cmap='gray')
```

```python
corners = cv2.goodFeaturesToTrack(gray,25,0.01,10)
corners = np.int0(corners)
```

```python
for i in corners:
    x,y = i.ravel()
    cv2.circle(image,(x,y),3,255,-1)

plt.imshow(image),plt.show()
```
