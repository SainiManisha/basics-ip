```python
import numpy as np
import cv2 #OpenCV-Python
%matplotlib inline
import matplotlib.pyplot as plt

print("OpenCV-Python Version %s" % cv2.__version__)
```

```python
#Import an image
image = cv2.imread('C:/Users/Manisha/Desktop/OpenCV/images/xyz.jpg')

image
```

```python
plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
```

```python
# Convert image color(BGR->Grayscale)
gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
# You can convert the image when calling cv2.imread()
# gray = cv2.imread('images/beaver.png', cv2.IMREAD_GRAYSCALE)

print(str(image.shape) + " => " + str(gray.shape))
plt.imshow(gray, cmap='gray')
```

```python
# SURF feature detector/descriptor
surf = cv2.xfeatures2d.SURF_create()
```

```python
# Find keypoints and descriptors directly
kp, des = surf.detectAndCompute(image,None)
```

```python
len(kp)
```

```python
img2 = cv2.drawKeypoints(image,kp,None,(255,0,0),4)
```
